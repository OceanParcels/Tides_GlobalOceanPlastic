We plot the mean separation (averaged over all particle pairs) of the GC+FES run as a function of time.

```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
File1 = '/science/projects/oceanparcels/output_data/data_Miriam/Results_TrackingGC.nc'
dataset1 = Dataset(File1)
time1 = dataset1.variables['time'][:]

File2 = '/science/projects/oceanparcels/output_data/data_Miriam/Results_TrackingGCFES.nc'
dataset2 = Dataset(File2)
time2 = dataset2.variables['time'][:]
```

```python
meanSep = np.load('/science/projects/oceanparcels/output_data/data_Miriam/MeanSeparationTime') 
meanSep = meanSep/1e3 # separation (mean of all pairs) as a function of time
time = time1[0,:]
years = time/31556926.0 # number of years that have passed since 2002-01-01
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6.5,4))
ax.loglog(years, meanSep, 'b')
ax.set_xlabel('Number of years since 2002-01-01', fontsize=12)
ax.set_xlim(left=0,right=13)
ax.set_xticks([1e-2,1e-1,1e0,1e1])
ax.set_xticklabels(['$10^{-2}$','$10^{-1}$','$10^{0}$','$10^{1}$'], fontsize=11.5)
ax.set_ylabel('Mean separation ($10^3$ km)', fontsize=12)
ax.set_yticks([1e-3,1e-2,1e-1,1e0])
ax.set_yticklabels(['$10^{-3}$','$10^{-2}$','$10^{-1}$','$10^{0}$'], fontsize=11.5)
ax.grid(True)
ax.set_title('Mean separation as a function of time', fontsize=13, fontweight='bold')
plt.tight_layout()
#plt.savefig('SeparationTime.pdf')
```
