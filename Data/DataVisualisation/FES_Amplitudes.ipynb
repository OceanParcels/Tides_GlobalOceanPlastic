# Visualising the FES amplitudes of the tidal constituents

*Author: Miriam Sterl*

We plot the eastward and northward amplitudes of the M$_2$, S$_2$, K$_1$ and O$_1$ tidal constituents from the FES data set for all locations on the globe.

```python
from netCDF4 import Dataset
import numpy as np
import numpy.ma as ma
import math
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm
import matplotlib as mpl
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.mpl.ticker as cticker
```

## Data sets

```python
File_M2_east = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/eastward_velocity/m2.nc'
dataset_M2_east = Dataset(File_M2_east)
Ua_M2 = dataset_M2_east.variables['Ua'][:]
lat_M2_east = dataset_M2_east.variables['lat'][:]
lon_M2_east = dataset_M2_east.variables['lon'][:]

File_M2_north = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/northward_velocity/m2.nc'
dataset_M2_north = Dataset(File_M2_north)
Va_M2 = dataset_M2_north.variables['Va'][:]
lat_M2_north = dataset_M2_north.variables['lat'][:]
lon_M2_north = dataset_M2_north.variables['lon'][:]

File_S2_east = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/eastward_velocity/s2.nc'
dataset_S2_east = Dataset(File_S2_east)
Ua_S2 = dataset_S2_east.variables['Ua'][:]
lat_S2_east = dataset_S2_east.variables['lat'][:]
lon_S2_east = dataset_S2_east.variables['lon'][:]

File_S2_north = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/northward_velocity/s2.nc'
dataset_S2_north = Dataset(File_S2_north)
Va_S2 = dataset_S2_north.variables['Va'][:]
lat_S2_north = dataset_S2_north.variables['lat'][:]
lon_S2_north = dataset_S2_north.variables['lon'][:]

File_K1_east = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/eastward_velocity/k1.nc'
dataset_K1_east = Dataset(File_K1_east)
Ua_K1 = dataset_K1_east.variables['Ua'][:]
lat_K1_east = dataset_K1_east.variables['lat'][:]
lon_K1_east = dataset_K1_east.variables['lon'][:]

File_K1_north = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/northward_velocity/k1.nc'
dataset_K1_north = Dataset(File_K1_north)
Va_K1 = dataset_K1_north.variables['Va'][:]
lat_K1_north = dataset_K1_north.variables['lat'][:]
lon_K1_north = dataset_K1_north.variables['lon'][:]

File_O1_east = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/eastward_velocity/o1.nc'
dataset_O1_east = Dataset(File_O1_east)
Ua_O1 = dataset_O1_east.variables['Ua'][:]
lat_O1_east = dataset_O1_east.variables['lat'][:]
lon_O1_east = dataset_O1_east.variables['lon'][:]

File_O1_north = '/science/projects/oceanparcels/input_data/DataPlasticTides/FES/northward_velocity/o1.nc'
dataset_O1_north = Dataset(File_O1_north)
Va_O1 = dataset_O1_north.variables['Va'][:]
lat_O1_north = dataset_O1_north.variables['lat'][:]
lon_O1_north = dataset_O1_north.variables['lon'][:]
```

## Eastward amplitudes

```python
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(2,2, figsize=(15,8), subplot_kw={'projection': ccrs.PlateCarree()})

Lon, Lat = np.meshgrid(lon_M2_east, lat_M2_east)
velplot = ax1.pcolormesh(Lon, Lat, Ua_M2, cmap='viridis', vmin=0, vmax=10)
ax1.coastlines()
ax1.add_feature(cfeature.LAND, zorder=0, edgecolor='black', facecolor=(0.6,0.6,0.6))
ax1.set_xticks([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], crs=ccrs.PlateCarree())
ax1.set_xticklabels([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], fontsize=12)
ax1.set_yticks([-90, -60, - 30, 0, 30, 60, 90], crs=ccrs.PlateCarree())
ax1.set_yticklabels([-90, -60, - 30, 0, 30, 60, 90], fontsize=12)
lon_formatter = cticker.LongitudeFormatter()
lat_formatter = cticker.LatitudeFormatter()
ax1.xaxis.set_major_formatter(lon_formatter)
ax1.yaxis.set_major_formatter(lat_formatter)
ax1.grid(linewidth=2, color='black', alpha=0.25, linestyle=':')
ax1.set_title('(a) M$_2$', fontsize = 14, fontweight = 'bold')

Lon, Lat = np.meshgrid(lon_S2_east, lat_S2_east)
velplot = ax2.pcolormesh(Lon, Lat, Ua_S2, cmap='viridis', vmin=0, vmax=10)
ax2.coastlines()
ax2.add_feature(cfeature.LAND, zorder=0, edgecolor='black', facecolor=(0.6,0.6,0.6))
ax2.set_xticks([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], crs=ccrs.PlateCarree())
ax2.set_xticklabels([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], fontsize=12)
ax2.set_yticks([-90, -60, - 30, 0, 30, 60, 90], crs=ccrs.PlateCarree())
ax2.set_yticklabels([-90, -60, - 30, 0, 30, 60, 90], fontsize=12)
lon_formatter = cticker.LongitudeFormatter()
lat_formatter = cticker.LatitudeFormatter()
ax2.xaxis.set_major_formatter(lon_formatter)
ax2.yaxis.set_major_formatter(lat_formatter)
ax2.grid(linewidth=2, color='black', alpha=0.25, linestyle=':')
ax2.set_title('(b) S$_2$', fontsize = 14, fontweight = 'bold')

Lon, Lat = np.meshgrid(lon_K1_east, lat_K1_east)
velplot = ax3.pcolormesh(Lon, Lat, Ua_K1, cmap='viridis', vmin=0, vmax=10)
ax3.coastlines()
ax3.add_feature(cfeature.LAND, zorder=0, edgecolor='black', facecolor=(0.6,0.6,0.6))
ax3.set_xticks([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], crs=ccrs.PlateCarree())
ax3.set_xticklabels([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], fontsize=12)
ax3.set_yticks([-90, -60, - 30, 0, 30, 60, 90], crs=ccrs.PlateCarree())
ax3.set_yticklabels([-90, -60, - 30, 0, 30, 60, 90], fontsize=12)
lon_formatter = cticker.LongitudeFormatter()
lat_formatter = cticker.LatitudeFormatter()
ax3.xaxis.set_major_formatter(lon_formatter)
ax3.yaxis.set_major_formatter(lat_formatter)
ax3.grid(linewidth=2, color='black', alpha=0.25, linestyle=':')
ax3.set_title('(c) K$_1$', fontsize = 14, fontweight = 'bold')

Lon, Lat = np.meshgrid(lon_O1_east, lat_O1_east)
velplot = ax4.pcolormesh(Lon, Lat, Ua_O1, cmap='viridis', vmin=0, vmax=10)
ax4.coastlines()
ax4.add_feature(cfeature.LAND, zorder=0, edgecolor='black', facecolor=(0.6,0.6,0.6))
ax4.set_xticks([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], crs=ccrs.PlateCarree())
ax4.set_xticklabels([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180], fontsize=12)
ax4.set_yticks([-90, -60, - 30, 0, 30, 60, 90], crs=ccrs.PlateCarree())
ax4.set_yticklabels([-90, -60, - 30, 0, 30, 60, 90], fontsize=12)
lon_formatter = cticker.LongitudeFormatter()
lat_formatter = cticker.LatitudeFormatter()
ax4.xaxis.set_major_formatter(lon_formatter)
ax4.yaxis.set_major_formatter(lat_formatter)
ax4.grid(linewidth=2, color='black', alpha=0.25, linestyle=':')
ax4.set_title('(d) O$_1$', fontsize = 14, fontweight = 'bold')

cbar_ax = fig.add_axes([1.01, 0.325, 0.01, 0.45])
cbar = fig.colorbar(velplot, cax=cbar_ax, extend='max')
cbar.set_label("Velocity (cm/s)", rotation=90, fontsize=13)
cbar.ax.tick_params(labelsize=13)
fig.suptitle('Eastward amplitudes', x=0.51,y=1, fontsize=18,fontweight='bold')
plt.tight_layout()
#plt.savefig('FES_Amplitudes_eastward', bbox_inches='tight')
```
